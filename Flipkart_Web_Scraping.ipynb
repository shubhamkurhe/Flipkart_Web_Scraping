### Lets create functions for each steps

```python
#Function to extract product title
def get_title(soup):
    
    try:
        title = new_soup.find('span', attrs = {'class':'B_NuCI'})
        
        title_string = title.text
        
    except AttributeError:
        title_string = ''
        
    return title_string
```

```python
# function to extract product price

def get_price(soup):
    
    try:
        price = new_soup.find('div', attrs = {'class':'_30jeq3 _16Jk6d'})
        
        price_string = price.text
        
    except AttributeError:
        price_string = ''
        
    return price_string
```

```python
#Function to extract product rating

def get_rating(soup):
    
    try:
        rating = new_soup.find('div', attrs = {'class': '_3LWZlK'}).text

    except AttributeError:
        rating = ''
        
    return rating
```

```python

# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36 Edg/115.0.1901.183',
       'Accept-Language':'en-US, en;=q=0.5'})

# The webpage URL
URL = 'https://www.flipkart.com/search?q=mobile+phones&otracker=search&otracker1=search&\
marketplace=FLIPKART&as-show=on&as=off&page=1'

# Webpage request
webpage = requests.get(url=URL, headers = HEADERS)

# Soup object containing all the data
soup = BeautifulSoup(webpage.content, 'html.parser')

# Fetch links as list of tags
links = soup.find_all('a', attrs = {'class':'_1fQZEK'})

# Store the links
links_list = []

# Loop for extracting links from Tag objects
# Loop through each link to get http url of each product
for link in links:
    links_list.append(link.get('href'))

d = {'title':[], 'price':[], 'rating':[]}


for link in links_list:

    new_webpage = requests.get('https://www.flipkart.com' + link, headers = HEADERS)

    new_soup = BeautifulSoup(new_webpage.content, 'html.parser')


    #Function calls to display all necessary product information

    d['title'].append(get_title(new_soup))
    d['price'].append(get_price(new_soup))
    d['rating'].append(get_rating(new_soup))

```

```python
d
```

```python
flipkart_df = pd.DataFrame.from_dict(d)
flipkart_df
```

```python
flipkart_df = pd.DataFrame.from_dict(d)
flipkart_df
```

```python
flipkart_df['title'].replace('', np.nan, inplace = True)
flipkart_df
```

```python
flipkart_df.dropna(subset = ['title'], inplace = True)
flipkart_df
```
